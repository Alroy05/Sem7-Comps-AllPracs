# Code

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
customer_data = pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
customer_data.shape
```

```python
customer_data.head(5)
```

```python
customer_data.describe()
```

```python
customer_data.info()
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
data
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
labels_=cluster.fit_predict(data)
```

```python
labels_
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```

# Adding Outliers

```python
# Create some outliers in the data
outliers = np.array([[700, 100], [800, 500], [750, 150]])  # Example outlier points
data = customer_data.iloc[:, 3:5].values
data_with_outliers = np.vstack((data, outliers))

data_with_outliers
```

```python
# Plot the dendrogram with outliers
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms (with outliers)")
dend = shc.dendrogram(shc.linkage(data_with_outliers, method='ward'))

plt.show()
```

```python
# Scatter plot with cluster labels (including outliers)
plt.figure(figsize=(10, 7))

# Plot the data points (excluding outliers)
plt.scatter(data[:, 0], data[:, 1], c=cluster.labels_[:data.shape[0]], cmap='rainbow', label='Data Points')

# Plot the outliers with a different marker and color
plt.scatter(data_with_outliers[-3:, 0], data_with_outliers[-3:, 1], c='red', marker='x', s=100, label='Outliers')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation with Outliers')
plt.legend()
plt.show()
```
